```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('profiles.csv')
```

```python
prof= df.copy().dropna(subset=['essay4'])
prof['essay4']= prof['essay4'].apply(lambda x: x.lower())
sinatraLovers= prof[prof['essay4'].str.contains('sinatra')].copy()
pinkfloydLovers= prof[prof['essay4'].str.contains('pink floyd')].copy()
beatlesLovers= prof[prof['essay4'].str.contains('beatles')].copy()
dylanLovers= prof[prof['essay4'].str.contains('dylan')].copy()
coldplayLovers= prof[prof['essay4'].str.contains('coldplay')].copy()
```

```python
fig, ax= plt.subplots(ncols=5, sharey=True,figsize=(15,8))
s=sns.countplot(data=sinatraLovers,x='drugs',ax=ax[0])
p=sns.countplot(data=pinkfloydLovers,x='drugs',ax=ax[1])
b=sns.countplot(data=beatlesLovers,x='drugs',ax=ax[2])
c=sns.countplot(data=coldplayLovers,x='drugs',ax=ax[3])
d=sns.countplot(data=dylanLovers,x='drugs',ax=ax[4])
c.set_xticklabels(list(c.get_xticklabels()), rotation=45)
s.set_xticklabels(list(s.get_xticklabels()), rotation=45)
b.set_xticklabels(list(b.get_xticklabels()), rotation=45)
p.set_xticklabels(list(p.get_xticklabels()), rotation=45)
d.set_xticklabels(list(d.get_xticklabels()), rotation=45)
c.set_title('ColdPlay')
s.set_title('Frank Sinatra')
p.set_title('Pink Floyd')
b.set_title('The Beatles')
d.set_title('Bob Dylan')
```

```python
#df['music']=df.dropna().apply(lambda x: str(x).split('music')[1].split('food'[0]))
#for x in df['essay4'].head(10).tolist():
    #print(x)
    
df['countwords']= df['essay0'].dropna().apply(lambda x:int(len(x.split())/20)*20)
df['counti']= df['essay0'].dropna().apply(lambda x:len(x.lower().split(' i ')))
```

```python
plt.figure(figsize=(8,4))
wordplot=sns.stripplot(data=df.dropna(subset=['countwords']),y='countwords',x='sex',jitter=0.05)
wordplot.set_title('Word count by sex')
wordplot.set_ylabel('Word count')
```

Surprisingly (i was surprised at least) men write more then women (i checked the self summary column as a reference) 

```python
plt.figure(figsize=(8,4))
Iplot=sns.stripplot(data=df.dropna(subset=['counti']),y='counti',x='income',jitter=0.05)
Iplot.set_title('i count by income')
Iplot.set_ylabel('I count')
Iplot.set_xticklabels(list(Iplot.get_xticklabels()), rotation=45)
plt.show()
```

people who prefer not to disclose their income write more I's and possibly are more self-centered

```python
plt.figure(figsize=(12,8))
wplot=sns.stripplot(data=df.dropna(subset=['countwords']),y='countwords',x='smokes',jitter=0.1)
wplot.set_title('word count by Smoking habits')
wplot.set_ylabel('Word count')
wplot.set_xticklabels(list(wplot.get_xticklabels()), rotation=45)
plt.show()
```

Smokers write less words about themselves

```python
sns.pairplot(data=df.dropna(),vars=['age','counti'],hue='sex',size=5)
```

there is no significant connection between age and I amount

```python
plt.figure(figsize=(15,4))
jobplot=sns.countplot(data=df,x='job',hue='sex')
jobplot.set_xticklabels(list(jobplot.get_xticklabels()), rotation=45)
plt.show()
```

The women are in more in executive jobs and in sales then their relative part of the population
